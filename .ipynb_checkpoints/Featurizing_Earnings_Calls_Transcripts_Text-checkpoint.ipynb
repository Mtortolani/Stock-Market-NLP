```python
!pip install PyPDF2
!pip install pdfplumber

import PyPDF2

class PdfCleaner:
  # read pdf files and extract texts
  def __init__ (self,pdf_file_path):
    import pdfplumber
    self.__path__ = pdf_file_path
    text = ""
    with pdfplumber.open(self.__path__) as pdf:
      for i in range(len(pdf.pages)):
        page = pdf.pages[i]
        text = text + str(page.extract_text())
    self.__original_text__ = text
    self.__cleaned_text__= self.clean_stopwords_punctuation()

  # Remove stopwords and punctuations from the original texts
  def clean_stopwords_punctuation(self):
    import gensim
    from gensim.parsing.preprocessing import remove_stopwords, STOPWORDS
    cleaned_text = remove_stopwords(self.__original_text__)
    punc = '''!()-[]{};:'"\,<>./?@#$%^&*_~©|'''
    for ele in cleaned_text:
      if ele in punc:
         cleaned_text = cleaned_text.replace(ele, "")
    return cleaned_text

  def lenBeforeClean(self):
    return len(self.__original_text__)

  def lenAfterClean(self):
    return len(self.__cleaned_text__)

  def print_originalText(self):
    print(self.__original_text__)

  def print_cleanedText(self):
    print(self.__cleaned_text__)


# Test
origin = PdfCleaner("APPL_2016Q3.pdf")
# origin.print_originalText()

# clean texts
print("\n\n------------Cleaned Texts-------------\n")
cleaned_text = origin.clean_stopwords_punctuation()
# origin.print_cleanedText()

# print length before and after cleaning
print("\n")
print("The length of texts before cleaning: "+str(origin.lenBeforeClean()))
print("The length of texts after cleaning: "+str(origin.lenAfterClean()))





from sklearn.feature_extraction.text import CountVectorizer 

''' Use countvectorizer to convert pdf text to a matrix of token counts '''
text = [cleaned_text]
vect = CountVectorizer()
vect.fit(text)
print("vocabulary: \n", vect.vocabulary_)


''' Use bag of words to count how many times each word appears in pdf '''
bag_of_words = vect.transform(text)
print(bag_of_words)

# number of times each unique word appears in this text
bag_of_words_arr = bag_of_words.toarray()
print(bag_of_words_arr)
feature_names = vect.get_feature_names()
print(feature_names)

''' Display occurrences in dataframe '''
import pandas as pd
quote_df = pd.DataFrame(bag_of_words_arr, columns = feature_names)
quote_df["APPL 2016 Q3 text"] = text

''' Convert bag of word values into TFIDF values (based on the number of occurrences for each word) '''
from sklearn.feature_extraction.text import TfidfTransformer
tfidfconverter = TfidfTransformer()
X = tfidfconverter.fit_transform(bag_of_words_arr).toarray()

print(X)

''' Display TFIDF values in dataframe '''
df = pd.DataFrame(X, columns = feature_names)
df["APPL 2016 Q3 text"] = text
```
